```python
import pandas as pd
import os 

files = os.listdir()
files = sorted(files)
```

```python
fp_dataframe = pd.DataFrame()
tp_dataframe = pd.DataFrame()
count = 1 
columns = []

for i in files:
    if "~" in i:
        print(i)
        
    elif ".xlsx" in i:
        if count == 1:
            columns.append(str(i))
            file_name = str(i).split()[0] + " sample ..."
            read_multi = pd.read_excel(i, sheet_name = file_name)
            read_multi = read_multi[read_multi["Resource"] == "Homo sapiens"]
            fp_dataframe[str(i).split()[0]] = read_multi["Mature 5'"].tolist()
            tp_dataframe[str(i).split()[0]] = read_multi["Mature 3'"].tolist()
            fp_dataframe.index = read_multi["Name"]
            tp_dataframe.index = read_multi["Name"]
            count += 1
            
        else:
            print(str(i))
            columns.append(str(i))
            file_name = str(i).split()[0] + " sample ..."
            read_multi = pd.read_excel(i, sheet_name = file_name)
            read_multi = read_multi[read_multi["Resource"] == "Homo sapiens"]
            five = read_multi[["Name","Mature 5'"]].set_index("Name")
            three = read_multi[["Name","Mature 3'"]].set_index("Name")
            fp_dataframe = pd.merge(fp_dataframe, five, how = "left", left_index = True, right_index = True)
            tp_dataframe = pd.merge(tp_dataframe, three, how = "left", left_index = True, right_index = True)
        
            print(fp_dataframe)
   
            
            
        
```

```python
fp_dataframe.columns = columns
tp_dataframe.columns = columns
columns_zelline = [0,0,0,5,5,5,9,9,9,16,16,16,26,26,26,36,36,36]

ad3 = fp_dataframe.iloc[:,:18]
ad2 = fp_dataframe.iloc[:,18:36]
eight = fp_dataframe.iloc[:,36:54]

ad3.columns = columns_zelline
ad2.columns = columns_zelline
eight.columns = columns_zelline 
```

```python
ad3 = ad3.reset_index()
ad2 = ad2.reset_index()
eight = eight.reset_index()
```

```python
gesamt = pd.concat([ad3,ad2,eight])
gesamt = gesamt.fillna(0)
gesamt = gesamt.groupby("Name").agg("mean")
```

```python
mean = gesamt.mean(axis = 1)
gesamt["mean"] = mean
gesamt = gesamt[gesamt["mean"] > 0]
gesamt.to_csv("miRNA_nc_multimappers_5p.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
five_p = pd.read_csv("miRNA_nc_multimappers_5p.csv").set_index("Name").drop("mean", axis = 1)
three_p = pd.read_csv("miRNA_nc_multimappers_3p.csv").set_index("Name").drop("mean", axis = 1)
five_p.columns = columns_zelline
three_p.columns = columns_zelline
```

```python
draw_mirna = ["mir-302a","mir-99a","let-7a-1","mir-25"]

def drawing(searched_mirna, five, three):
    five_p_mirna = five[five.index.isin(searched_mirna)]
    three_p_mirna = three[three.index.isin(searched_mirna)]
    
    if (five_p_mirna.shape[0] > 0 and three_p_mirna.shape[0] > 0):
        fig, ax = plt.subplots(ncols = 2, nrows = 1, figsize = (10,5))
        sns.lineplot(data = five_p_mirna.T, ax = ax[0])
        sns.lineplot(data = three_p_mirna.T, ax = ax[1])
        ax[0].set_title("5' mature miRNA")
        ax[1].set_title("3' mature miRNA")
        plt.show()
        
    elif (five_p_mirna.shape[0] == 0 and three_p_mirna.shape[0] > 0):
        fig, ax = plt.subplots(ncols = 1, nrows = 1, figsize = (10,5))
        sns.lineplot(data = three_p_mirna.T)
        plt.show()
        
    elif (five_p_mirna.shape[0] > 0 and three_p_mirna.shape[0] == 0):
        fig, ax = plt.subplots(ncols = 1, nrows = 1, figsize = (10,5))
        sns.lineplot(data = five_p_mirna.T)
        plt.show()
        
    else:
        print("something went wrong")
        
        
    
        
        
drawing(draw_mirna, five_p, three_p)
```
