```python
# a script to collect the sample tpm expression
# mapped to grch.38.p7 from NCBI using the CLC genome workbench
import pandas as pd
import seaborn as sns
import numpy as np

tpm = pd.read_csv("../files/tpm_expressed_mrnas/tpm_for_all_sample.csv")
tpm = tpm.set_index("gene_name")
```

```python
def draw_function(table, gene_name):
    """drawing of the table"""
    table_gene = table[table.index == gene_name]
    columns = [[i]*3 for i in ["D0","D5","D9","D16","D26","D36"]]
    columns = sum(columns, [])
    print(columns)
    ad3 = table_gene.iloc[:,0:18]
    ad2 = table_gene.iloc[:,18:36]
    eight = table_gene.iloc[:,36:]
    ad3.columns = columns
    ad2.columns = columns
    eight.columns = columns
    
    ad3.T.plot()
    ad2.T.plot()
    eight.T.plot()
    
draw_function(tpm, "EXOC1")
    
```
