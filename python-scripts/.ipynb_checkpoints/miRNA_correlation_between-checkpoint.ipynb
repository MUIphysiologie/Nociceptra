```python
# a basic script more miRNA and mRNA correlation analysis between cell lines
# 3d plot of correlation between the cell lines
#extra script not in the manuscript
# can be used to access the equality of the three cell lines
import pandas as pd
import matplotlib.pyplot as plt

ad2 = pd.read_csv("../files/normalized_counts_DESEQ2/AD2_miRNAs_timeline_ltr-results-with-normalized.csv")
ad3 = pd.read_csv("../files/normalized_counts_DESEQ2/AD3_miRNAs_timeline_ltr-results-with-normalized.csv")
eigth = pd.read_csv("../files/normalized_counts_DESEQ2/840_miRNAs_timeline_ltr-results-with-normalized.csv")
```

```python
# for the target score the baseline expression of each individual miRNA needs to be accessed and stored
# this was done on the previously loaded unfiltered AD2, AD3, 840 expression datasets 

baseline_expression = pd.concat([ad2,ad3, eigth])
baseline_expression = baseline_expression.reset_index()
baseline_expression = pd.DataFrame(baseline_expression.groupby("Row.names")["baseMean"].agg("mean"))
baseline_expression.sort_values("baseMean", ascending = False)
#baseline_expression.to_csv("mean_baseline_expression_miRNA.csv")
```

```python
# correlation anaylsis between each individual miRNA or gene between the cell lines
# Pearson distribution was used also Spearman could be used 
from scipy import stats
def correlation_cell(df1, df2, df3):
    corr = {"mirna":[], "ad2_ad3":[],"ad2_eight":[],"ad3_eight":[]}
    for i in set(df1["Row.names"]):
        ad2 = df1[df1["Row.names"]==i].set_index("Row.names").iloc[:,8:].values
        ad3 = df2[df2["Row.names"]==i].set_index("Row.names").iloc[:,8:].values
        eight = df3[df3["Row.names"]==i].set_index("Row.names").iloc[:,8:].values
        ad2_ad3 = stats.pearsonr(ad2[0], ad3[0])
        ad2_eight = stats.pearsonr(ad2[0], eight[0])
        ad3_eight= stats.pearsonr(ad3[0], eight[0])
        corr["mirna"].append(i)
        corr["ad2_ad3"].append(ad2_ad3[0])
        corr["ad2_eight"].append(ad2_eight[0])
        corr["ad3_eight"].append(ad3_eight[0])
    return corr

correlation = pd.DataFrame(correlation_cell(ad2,ad3,eigth))
```

```python
# get meaningfull information about the number of miRNA or genes highly correlated 
# draw the violinplot for thee mean correlation between the cell lines

from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['svg.fonttype'] = 'none'
correlation = correlation.dropna()
print(correlation.describe())
corr = correlation[(correlation["ad2_ad3"] > 0.5) & (correlation["ad2_eight"] >0.5) & (correlation["ad3_eight"] >0.5)]
corr2 = correlation[~((correlation["ad2_ad3"] > 0.5) & (correlation["ad2_eight"] >0.5) & (correlation["ad3_eight"] >0.5))]
fig,ax = plt.subplots(figsize = (5,5))
sns.despine()
sns.violinplot(data = correlation, palette = "tab20b")
#plt.savefig("miRNA_correlation_violinplots.svg", quality = 90, dpi = 600, transparent = True, bbox_inches = "tight")
```

```python
#Plot the 3d Scatter plot between all the three cell lines to idnetify that the same genes between the cell lines comparison
# are indeed highly correlated


fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(corr2["ad2_ad3"],corr2["ad2_eight"],corr2["ad3_eight"], color = "black", alpha = 0.2)
ax.scatter(corr["ad2_ad3"],corr["ad2_eight"],corr["ad3_eight"], color = "red", alpha = 0.7)
ax.grid(False)
ax.set_xlabel("r (AD2-AD3)")
ax.set_ylabel("r (AD2-840)")
ax.set_zlabel("r (AD3-840)")
ax.set_xticks([-1,0,1])
ax.set_yticks([-1,0,1])
ax.set_zticks([-1,0,1])
ax.view_init(30,230)
#plt.savefig("miRNA_3d_scatter_all_correlation.svg", quality = 90, dpi = 600, transparent = True, bbox_inches = "tight")
```
