```python
def disease_genes(all_diseases, queried_disease):
    """query disease genes for the choosen disease such as chronic pain 
    returns the list of genes associated with the disease"""
    disease_genes = set(all_diseases[all_diseases["diseaseName"].str.contains(queried_disease, case=False)]["geneSymbol"].tolist())
    return list(disease_genes)


genes_disease = disease_genes(diseases, "Chronic Pain")
genes_disease
```

```python
# Interactions of the disease associated genes
```

```python
def get_interaction(genes, string):
    string_network = string[string["preferred_name_x"].isin(genes)]
    string_end = string_network[string_network["preferred_name_y"].isin(genes)]
    string_end = string_end.drop(["Unnamed: 0", "score"], axis=1)
    return string_end


links = get_interaction(load_pain_t, string_interaction_high)
links.columns = ["source", "target", "value"]
links
```

```python
# check all nodes that are in the links list
nodes = list(set(links["source"].tolist() + links["target"].tolist())) # remove unwanted duplicates
#retrieve the supercluster stages for alle of the genes in the list
gesamt = supercluster[supercluster["external_gene_name"].isin(nodes)]
gesamt = gesamt.reset_index()

#shallow copy of gesamt 
node = gesamt
node.columns = ["index", "gene_name", "supercluster_gene"] #rename columns
links = links[(links["source"].isin(node["gene_name"]))& (links["target"].isin(node["gene_name"]))]
node = node[node["gene_name"].isin(links["source"].tolist() +links["target"].tolist())]
#create a network index for the holoviews plot
links_index = pd.merge(links, 
                       node,
                       how="left",
                       left_on="source",
                       right_on="gene_name")
links_index = pd.merge(links_index,
                       node,
                       how="left",
                       left_on="target",
                       right_on="gene_name")
print(links_index)

#set the index and make and hv dataset out of the data
links_index = links_index[["index_x", "index_y", "value", "supercluster_gene_x"]]
links_index.columns = ["source", "target", "value", "edges"]
links_index = links_index[~links_index[['source', 'target']].apply(frozenset, 1).duplicated()]
nodes = hv.Dataset(pd.DataFrame(node.sort_values("supercluster_gene")), "index")
links_index
```

```python
#plot the chord plot --> use bokeh or matplotlib extensions
hv.extension("matplotlib")
chord = hv.Chord((links_index, nodes)).opts(
    opts.Chord(cmap='tab20b',
               edge_cmap='tab20b',
               labels=dim("gene_name").str(),
               node_color=dim('supercluster_gene').str(),
               fig_size=450,
               node_size=1,
               fontsize={"labels": 4}))
#hv.save(chord, "neuropathic_pain_plot.svg", dpi=500) 
#save the plot by uncommenting the line above
```

```python
def gene_set_kegg_enrichment(genes_list, genes_query, genes_baseline):
    """ Here we write a function to define contigency table of observed genes and expected genes"""
    genes_diagram = genes_baseline.set_index("Unnamed: 0.1").drop("Unnamed: 0",
                                                                  axis=1)
    genes_expected = genes_query[genes_query["external_gene_name"].isin(genes_list)]
    genes_expected = pd.DataFrame(genes_expected["supercluster_gene"].value_counts())
    
    gesamt = pd.merge(genes_diagram,
                      genes_expected,
                      how="left",
                      left_index=True,
                      right_index=True)
    print(gesamt)
    gesamt = gesamt.fillna(int(0))
    table = sm.stats.Table(gesamt)
    rslt = table.test_nominal_association()
    enrichments_residuals = pd.DataFrame(table.resid_pearson)
    enrichments_residuals.columns = ["observed_genes", "expected_genes"]
    p_value = rslt.pvalue
    print(p_value)
    statistics = enrichments_residuals.iloc[:, 1].reset_index()
    return statistics


statistics = gene_set_kegg_enrichment(load_pain_t, super_cluster_statistics,
                                      super_cluster_counts)

# put the statistics into a dataframe and draw a vertical barplot to show enrichments or
#depletions
stats = pd.DataFrame(statistics).set_index("Unnamed: 0.1")
fig, ax = plt.subplots(figsize=(5, 5))
sns.despine
sns.barplot(data=stats.T, orient="h")
plt.xlim(-3.5, 3.5)
plt.xlabel("Pearson Residuals", fontsize=15)
plt.ylabel("")
#plt.savefig("neuropathic_pain_pearson.svg",
 #           quality=90,
 #           transparent=True,
 #           bbox_inches="tight")
```
